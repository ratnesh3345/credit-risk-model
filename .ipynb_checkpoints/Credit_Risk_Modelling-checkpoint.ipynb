### Importing Libraries and Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df_loan= pd.read_csv("loans.csv")
df_customers = pd.read_csv("customers.csv")
df_bureau_data = pd.read_csv("bureau_data.csv")
```

```python
df = pd.merge(df, df_bureau_data, on="cust_id")
df.head()
```

```python
df.info()
```

```python
#Converting Our Target Variable from Boolean to Int
df["default"] = df["default"].astype(int)
```

### Creating a Train Test Split to avoid Data Leakage

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["default"])
y = df["default"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
df_train = pd.concat([X_train, y_train], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)
```

```python
#scaling the data to calculate VIF

X_train = df_train_3.drop(columns=["default"])
y_train = df_train_3["default"]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
cols_to_scale = ['age', 'number_of_dependants', 'years_at_current_address',
       'sanction_amount', 'processing_fee', 'gst', 'net_disbursement',
       'loan_tenure_months', 'principal_outstanding',
       'bank_balance_at_application', 'number_of_open_accounts',
       'number_of_closed_accounts', 'enquiry_count',
       'credit_utilization_ratio', 'loan_to_income', 'deliquent_ratio',
       'avg_dpd_per_delinquency']

X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_train.head()
```

```python
#for test dataset
#scaling the data to calculate VIF

X_test = df_test_3.drop(columns=["default"])
y_test = df_test_3["default"]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
cols_to_scale = ['age', 'number_of_dependants', 'years_at_current_address',
       'sanction_amount', 'processing_fee', 'gst', 'net_disbursement',
       'loan_tenure_months', 'principal_outstanding',
       'bank_balance_at_application', 'number_of_open_accounts',
       'number_of_closed_accounts', 'enquiry_count',
       'credit_utilization_ratio', 'loan_to_income', 'deliquent_ratio',
       'avg_dpd_per_delinquency']

X_test[cols_to_scale] = scaler.fit_transform(X_test[cols_to_scale])
X_test.describe()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(data):
    vif_df = pd.DataFrame()
    vif_df["columns"] = data.columns
    vif_df["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    return vif_df
```

```python
calculate_vif(X_train[cols_to_scale])
```
